```python
covariates = ['educ', 'exper', 'female', 'married']
```

```python
from sklearn.model_selection import KFold
(train, test), _, _ = KFold(n_splits=3).split(df)
train_df = df.loc[train]
test_df = df.loc[test]
```

```python

import numpy as np
from scipy.special import erf


def gaussian_cdf(z):
    return (1 + erf(z / np.sqrt(2))) / 2


def gaussian_pdf(z):
    return np.exp(-z ** 2 / 2) / np.sqrt(2 * np.pi)


def uniform_pdf(z):
    return np.abs(z) < 0.5


def epanechnikov_pdf(z):
    return 0.75 * (1 - z ** 2) * (np.abs(z) <= 1)
```

```python
train_y = train_df['wage'].values
train_Z = train_df[covariates].values
n, d = train_Z.shape

test_y = test_df['wage'].values
test_Z = test_df[covariates].values
```

```python
kernels = []
for i, covariate in enumerate(covariates):
    if len(np.unique(train_Z[:, i])) < 0.02 * n:
        print(f'We use uniform kernel for {covariate}')
        kernels.append(uniform_pdf)
    else:
        print(f'We use a continuous kernel for {covariate}')
        kernels.append(gaussian_pdf)
```

```python
h = 0.2

weight = 1
for i, kernel in enumerate(kernels):
    h_ = 1.06 * np.std(train_Z[:, i]) * np.power(n, -1/(2*2+2)) if kernel != uniform_pdf else 1
    print(h_)
    weight *= kernel((test_Z[None, :, i] - train_Z[:, None, i]) / h_) 

test_y_pred = np.average(train_y[:, None] * weight, axis=0) / np.average(weight, axis=0)
```

```python
np.mean((test_y - test_y_pred) ** 2)
```

```python
def kernel_regression_estimator(train_df, test_df, covariates, target='wage'):
    if not isinstance(covariates, dict):
        kernel_mapping = {}
        for i, covariate in enumerate(covariates):
            if len(np.unique(train_Z[:, i])) < 0.02 * n:
                kernels[covariate] = uniform_pdf
            else:
                h = 1.06 * np.std(train_Z[:, i]) * np.power(n, -1/5)
                kernels[covariate] = lambda z: epanechnikov_pdf(z/h)
    else:
        kernel_mapping = covariates
    
    covariates = list(kernel_mapping.keys())
    
    train_y = train_df[target]
    train_Z = train_df[covariates]

    test_Z = test_df[covariates]
    
    weight = 1
    for covariate, kernel in kernel_mapping.items():
        weight *= kernel((test_Z[covariate].values[None, :] - train_Z[covariate].values[:, None]))
        # print(kernel((test_Z[covariate].values[None, :] - train_Z[covariate].values[:, None])))
        
    # print(np.average(weight, axis=0))

    test_y_pred = np.average(train_y.values[:, None] * weight, axis=0) / np.average(weight, axis=0)
    return test_y_pred
```

```python
educ_hs = np.linspace(0.2, 3, 7) * np.std(train_df['educ']) * np.power(len(train_df), -1/5)
exper_hs = np.linspace(0.2, 3, 7) * np.std(train_df['exper']) * np.power(len(train_df), -1/5)

mise = np.zeros((len(educ_hs), len(exper_hs)))
for i, educ_h in enumerate(educ_hs):
    for j, exper_h in enumerate(exper_hs):
        test_y_pred = kernel_regression_estimator(train_df, test_df, {
            'educ': lambda z: gaussian_pdf(z/educ_h),
            'exper': lambda z: gaussian_pdf(z/exper_h),
            'female': uniform_pdf,
            'married': uniform_pdf,
        })
        mise[i, j] = np.mean((test_df['wage'] - test_y_pred) ** 2)
```
